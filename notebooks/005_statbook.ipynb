```python
import numpy as np
import torch
import os
```

```python
# for each child directory in /scratch/f006pq6/projects/asr-grad-reconstruction/logging/0s-1s, load the x_param_last.pt files. print out the keys() ot the file.

base_path = '/scratch/f006pq6/projects/asr-grad-reconstruction/logging/2s-3s'
# list child dir of base_path
all_mae = []
all_time = []
for child_dir in os.listdir(base_path):
    child_path = os.path.join(base_path, child_dir)
    if os.path.isdir(child_path):
        # load the x_param_last.pt file
        file_path = os.path.join(child_path, 'x_param_last.pt')
        if os.path.isfile(file_path):
            checkpoint = torch.load(file_path)
            
            # check point is dictonary with keys 'x_param', 'time', 'inputs', 'targets', 'transcript' 
            # now printe transcript
            print(checkpoint['transcript'])
            x_param = checkpoint['x_param']
            inputs = checkpoint['inputs'] 
            mae = torch.mean(torch.abs(x_param - inputs))
            all_mae.append(mae.item())
            all_time.append(checkpoint['time'])
            print(f'MAE: {mae.item()}')
        else:
            print(f'File {file_path} does not exist.')

print('mae mean, std:', np.mean(all_mae), np.std(all_mae))  # print mean and std of mae
print('time mean, std:', np.mean(all_time), np.std(all_time))  # print mean and std of time
```
